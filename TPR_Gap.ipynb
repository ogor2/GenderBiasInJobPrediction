```python
import pandas as pd
```

```python
df_lr = pd.read_csv("lr_predictions.csv")
df_bert = pd.read_csv("bert_predictions.csv")
```

```python
def calculate_bias(df, target_label):
    
    true_females = df[(df['true_label'] == target_label) & (df['gender'] == 1)]
    true_males = df[(df['true_label'] == target_label) & (df['gender'] == 0)]
    
    if len(true_females) > 0:
        correct_females = true_females[true_females['predicted_label'] == target_label]
        tpr_female = len(correct_females) / len(true_females)
    else:
        tpr_female = 0

    if len(true_males) > 0:
        correct_males = true_males[true_males['predicted_label'] == target_label]
        tpr_male = len(correct_males) / len(true_males)
    else:
        tpr_male = 0

    return (target_label, tpr_male - tpr_female, tpr_male, tpr_female)
```

## TPR Gap Logistic Regression
Below is the calculated true positive rate for both male and female of LR. The Gap is calculated by male - female.

```python
df_lr = df_lr.drop(columns = ['Unnamed: 0'], axis = 1)
df_lr = df_lr.rename(columns = {'y_true' : 'true_label', 'y_pred_lr': 'predicted_label'})
```

```python
df_lr
```

```python
for i in range(0, 28):
    target, gap, tpr_m, tpr_f = calculate_bias(df_lr, i)
    print(f"Target Label: {target}, TPR Gap: {gap}, TPR Male: {tpr_m}, TPR Female: {tpr_f} \n")
```
