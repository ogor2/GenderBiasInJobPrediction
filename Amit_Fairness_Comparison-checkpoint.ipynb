```python
import pandas as pd
from sklearn.metrics import accuracy_score, f1_score

# File names
LR_FILE = "lr_predictions.csv"
BERT_FILE = "bert_predictions.csv"
```

```python
# Logistic Regression predictions
lr = pd.read_csv(LR_FILE)

# id' column based on row index so we can match with BERT later
lr["id"] = lr.index

print("LR predictions (first 5 rows):")
display(lr.head())
print("Columns:", lr.columns.tolist())
print("Number of rows:", len(lr))
```

```python
# Overall performance for Logistic Regression
y_true = lr["y_true"]
y_lr   = lr["y_pred_lr"]

acc_lr = accuracy_score(y_true, y_lr)
f1_lr  = f1_score(y_true, y_lr, average="macro")

print("=== Overall Performance: Logistic Regression ===")
print(f"Accuracy : {acc_lr:.4f}")
print(f"Macro F1 : {f1_lr:.4f}")
```

```python
genders = lr["gender"].unique()
print("Gender groups in data:", genders)

results_lr = []

for g in genders:
    group = lr[lr["gender"] == g]
    y_true_g = group["y_true"]
    y_lr_g   = group["y_pred_lr"]

    acc_lr_g = accuracy_score(y_true_g, y_lr_g)
    f1_lr_g  = f1_score(y_true_g, y_lr_g, average="macro")

    results_lr.append({
        "gender": g,
        "acc_lr": acc_lr_g,
        "f1_lr":  f1_lr_g
    })

results_lr_df = pd.DataFrame(results_lr)
print("=== Logistic Regression performance by gender ===")
display(results_lr_df)
```

```python
acc_gap_lr = results_lr_df["acc_lr"].max() - results_lr_df["acc_lr"].min()
f1_gap_lr  = results_lr_df["f1_lr"].max()  - results_lr_df["f1_lr"].min()

print("=== Fairness gaps for Logistic Regression ===")
print(f"Accuracy gap between genders: {acc_gap_lr:.4f}")
print(f"F1 gap between genders      : {f1_gap_lr:.4f}")
```

```python
summary_lr = pd.DataFrame({
    "Model": ["Logistic Regression"],
    "Overall Accuracy": [acc_lr],
    "Overall F1 (macro)": [f1_lr],
    "Accuracy Gap (gender)": [acc_gap_lr],
    "F1 Gap (gender)": [f1_gap_lr]
})

print("=== Logistic Regression summary (use this in report/slides) ===")
display(summary_lr)

summary_lr.to_csv("fairness_summary_lr_amit.csv", index=False)
print("Saved LR summary as fairness_summary_lr_amit.csv")
```

## Fairness Evaluation of Logistic Regression (Amit Bhatt)

For our baseline model, we used a Logistic Regression classifier trained on the Bias in Bios dataset. On the held-out test set, this model achieved an overall accuracy of **0.82** and a macro F1-score of **0.76**.

To evaluate fairness, I computed the accuracy and macro F1-score separately for each gender group and then measured the **gap** between the best and worst group. The **accuracy gap between genders** was **0.013**, and the **F1-score gap** was **0.023**. These values are relatively small, which means the Logistic Regression model performs fairly similarly across gender groups, although a small difference in performance still exists.

This baseline will be used later to compare against the BERT classifier, both in terms of overall performance and fairness across gender.


```python
# Loading BERT predictions
bert = pd.read_csv("bert_predictions.csv")

# Drop old index column if there is one
if "Unnamed: 0" in bert.columns:
    bert = bert.drop(columns=["Unnamed: 0"])

print("BERT predictions (first 5 rows):")
display(bert.head())
print("Columns:", bert.columns.tolist())
print("Number of rows:", len(bert))
```

```python
# Making column names consistent with LR
bert = bert.rename(columns={
    "true_label": "y_true",
    "predicted_label": "y_pred_bert"
})

print("After renaming columns:")
display(bert.head())
print("Columns:", bert.columns.tolist())
```

```python
# Overall performance for BERT
y_true_bert = bert["y_true"]
y_bert      = bert["y_pred_bert"]

acc_bert = accuracy_score(y_true_bert, y_bert)
f1_bert  = f1_score(y_true_bert, y_bert, average="macro")

print("=== Overall Performance: BERT ===")
print(f"Accuracy : {acc_bert:.4f}")
print(f"Macro F1 : {f1_bert:.4f}")
```

```python
genders_bert = bert["gender"].unique()
print("Gender groups in BERT data:", genders_bert)

results_bert = []

for g in genders_bert:
    group = bert[bert["gender"] == g]
    y_true_g = group["y_true"]
    y_bert_g = group["y_pred_bert"]

    acc_bert_g = accuracy_score(y_true_g, y_bert_g)
    f1_bert_g  = f1_score(y_true_g, y_bert_g, average="macro")

    results_bert.append({
        "gender": g,
        "acc_bert": acc_bert_g,
        "f1_bert":  f1_bert_g
    })

results_bert_df = pd.DataFrame(results_bert)
print("=== BERT performance by gender ===")
display(results_bert_df)
```

```python
acc_gap_bert = results_bert_df["acc_bert"].max() - results_bert_df["acc_bert"].min()
f1_gap_bert  = results_bert_df["f1_bert"].max()  - results_bert_df["f1_bert"].min()

print("=== Fairness gaps for BERT ===")
print(f"Accuracy gap between genders: {acc_gap_bert:.4f}")
print(f"F1 gap between genders      : {f1_gap_bert:.4f}")
```

```python
summary_bert = pd.DataFrame({
    "Model": ["BERT"],
    "Overall Accuracy": [acc_bert],
    "Overall F1 (macro)": [f1_bert],
    "Accuracy Gap (gender)": [acc_gap_bert],
    "F1 Gap (gender)": [f1_gap_bert]
})

print("=== BERT summary (for report/slides) ===")
display(summary_bert)

summary_bert.to_csv("fairness_summary_bert_amit.csv", index=False)
print("Saved BERT summary as fairness_summary_bert_amit.csv")
```

```python
summary_all = pd.DataFrame({
    "Model": ["Logistic Regression", "BERT"],
    "Overall Accuracy": [acc_lr, acc_bert],
    "Overall F1 (macro)": [f1_lr, f1_bert],
    "Accuracy Gap (gender)": [acc_gap_lr, acc_gap_bert],
    "F1 Gap (gender)": [f1_gap_lr, f1_gap_bert]
})

print("=== Final Comparison: Logistic Regression vs BERT ===")
display(summary_all)

summary_all.to_csv("fairness_summary_lr_vs_bert_amit.csv", index=False)
print("Saved comparison summary as fairness_summary_lr_vs_bert_amit.csv")
```

### Fairness Evaluation and Model Comparison (Amit Bhatt)

In this section, I evaluate how fair our models are with respect to gender and compare the Logistic Regression baseline to the BERT classifier.

#### Logistic Regression Fairness

Our baseline model is a Logistic Regression classifier trained on the Bias in Bios dataset. On the held-out test set, this model achieved an overall accuracy of **0.82** and a macro F1-score of **0.76**.

To study fairness, I computed accuracy and macro F1-score separately for each gender group and then measured the gap between the best and worst group. For Logistic Regression, the **accuracy gap between genders** is **0.013**, and the **F1-score gap** is **0.023**. These gaps are relatively small, which suggests that the baseline model behaves similarly across genders, but there is still a small difference in performance between groups.

#### BERT Fairness

Next, I evaluated a BERT-based classifier on the same test set. The BERT model achieved an overall accuracy of **0.86** and a macro F1-score of **0.81**, which is higher than the Logistic Regression baseline.

Using the same fairness procedure, the **accuracy gap between genders** for BERT is **0.009**, and the **F1-score gap** is **0.020**. These values are also small and slightly lower than the gaps for Logistic Regression, meaning BERT is at least as fair and slightly more balanced across gender groups.

#### Comparison Between Logistic Regression and BERT

Table X summarizes the overall performance and fairness metrics for both models. BERT clearly improves predictive performance: accuracy increases from **0.82** (Logistic Regression) to **0.86**, and macro F1 increases from **0.76** to **0.81**. At the same time, BERT slightly reduces the fairness gaps: the accuracy gap drops from **0.013** to **0.009**, and the F1 gap decreases from **0.023** to **0.020**.

Overall, BERT is the better model for this task because it provides higher accuracy and a small improvement in fairness across gender. However, the gaps are not zero for either model, so some residual bias remains even after switching to a more powerful classifier.
